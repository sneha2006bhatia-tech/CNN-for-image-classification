```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN,self).__init__()
        self.convo_layer=nn.Sequential(
            nn.Conv2d(1,32,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32,64,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc_layer=nn.Sequential(
            nn.Linear(7*7*64,128),
            nn.ReLU(),
            nn.Linear(128,10)
        )
    def forward(self,x):
        x=self.convo_layer(x)
        x=x.view(x.size(0),-1)
        x=self.fc_layer(x)
        return x
```
